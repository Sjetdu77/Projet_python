# L'évolution des positions des personnages de _Super Smash Bros. Brawl_

#### _Super Smash Bros._ est un crossover sous forme d'une série de jeux de combat développé par _HAL Laboratory_, puis par _Sora_, édité par _Nintendo_ et toujours réalisé par Masahiro Sakurai. L'objectif est de battre son adversaire, non pas en lui retirant toute sa vie, mais en l'éjectant de l'arène. _Super Smash Bros._ est conçu de manière à ce que n'importe qui peut battre n'importe quel autre adversaire, même si ce dernier y joue depuis des décennies. L'opus _Brawl_, sorti en 2008 sur Wii, est l'apothéose de cet objectif. Il a pu y avoir une compétition, mais elle a été laborieuse, et les différences de puissance entre les personnages ont fait partie des obstacles. Comment est-ce que les personnages ont évolué dans ce jeu ? C'est ce qu'on va voir.

#### Avant de se poser la question de comment est-ce que les personnages ont évolué dans ce jeu, il faut d'abord savoir combien y a-t-il de personnages, mais grâce aux _tier lists_. Une _tier list_ est un classement concernant tous les personnages jouables d'un jeu vidéo par ordre de viabilité. Ces _tier lists_ ne sont valables qu'à haut niveau, donc dans une compétition uniquement. Les classements sont effectués par catégories, nommés _tiers_, qui commencent le plus souvent par S, puis par A pour continuer par ordre alphabétique. Il existe quelques exceptions à la règles : quand un personnage est beaucoup plus fort que les autres et qu'il faut créer un tier SS rien que pour lui, ce qui est le cas dans notre jeu. Voyons donc le nombre de personnages par catégorie.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from seaborn.categorical import boxplot as bxp
from datetime import datetime as dt
from random import randrange as rdrg

plt.figure(figsize=(12,6))
ax = plt.subplot()

tl = pd.read_excel('.\\data\\TL Brawl.xls', index_col=0)

g = sbn.countplot(x='Tier', data=tl, order=["SS", "S", "A+", "A-", "B", "C+", "C", "C-", "D", "E", "F"])

plt.show()
```

#### On peut déjà voir que les tiers ne sont pas croissants ni en cloche, mais aussi que ça peut donner une idée de combien est-ce qu'il y a de personnages. On peut estimer qu'il y a aux alentours de 35 personnages, 37 pour être plus précis. Donc, si on veut voir la progression de tous les personnages par tier lists, voici ce que cela donne.

```python
plt.figure(figsize=(12,6))
ax = plt.subplot()

tl = pd.read_excel('.\\data\\TL Brawl.xls', index_col=0)

tl = tl.T.iloc[:8]
ax.plot(tl, 'o')
ax.set_ylim(39, 0)
plt.xlabel("Tier Lists")
plt.ylabel("Position")

plt.show()
```

#### On remarque donc qu'on ne peut pas interpréter la progression de chaque personnage en fonction des _tier lists_. Cependant, on peut interpréter quelque chose de ce graphique : quand se sont passés les différents _tier lists_. En effet, chaque ligne de points correspond à une date de création d'une _tier list_. On remarque que le premier a été créé fin 2008, qu'il y ait eu deux _tier lists_ en 2009 et en 2010, pour arriver à une _tier list_ par an jusqu'à début 2013.

#### _Super Smash Bros._ étant un crossover, chaque personnage représente son univers. On peut donc regarder quels univers sont les plus représentés de _Super Smash Bros. Brawl_, et voici le résultat.

```python
plt.figure(figsize=(12,6))
ax = plt.subplot()

tl = pd.read_excel('.\\data\\TL Brawl.xls', index_col=0)

g = sbn.countplot(x='Univers', data=tl)

plt.show()
```

#### On remarque que _Mario_, _The Legend of Zelda_ et _Pokémon_ sont les trois univers les plus représentés du jeu, ce qui est normal au vue de l'éditeur. Maintenant, il faut regarder quel univers allons-nous regarder en fonction des variations de positions.

```python
plt.figure(figsize=(12,6))
ax = plt.subplot()

tl = pd.read_excel('.\\data\\TL Brawl.xls', index_col=0)

tl = tl[(tl['Univers'] == 'Mario') | (tl['Univers'] == 'Pokémon') | (tl['Univers'] == 'Zelda')]
g = bxp(x="Univers", y=dt(2013, 4, 25), data=tl)
plt.ylabel("Position")
ax.set_ylim(39, 0)

plt.show()
```

#### On voit que l'univers _Mario_ a le plus de variations dans les positions des personnages, donc on va prendre les personnages de cet univers précis. Maintenant, il est temps de répondre à la question de comment ont évolué les personnages du jeu. Voici donc le graphique montrant la progression des personnages.

```python
plt.figure(figsize=(12,6))
ax = plt.subplot()

tl = pd.read_excel('.\\data\\TL Brawl.xls', index_col=0)

tl = tl[tl['Univers'] == 'Mario']
tl = tl.T.iloc[:8]
col = tl.columns
ax.plot(tl, 'o-')
ax.set_ylim(39, 0)
plt.xlabel("Tier lists")
plt.ylabel("Position")
plt.legend(col)

plt.show()
```

#### Nous avons donc les huit personnages qui représentent l'univers _Mario_, et chacun a eu une évolution dans la _tier list_ différente : Yoshi a globalement progressé ; Wario a toujours été dans les classements assez élevés avec des pics de montée ; Donkey Kong a régressé au fur et à mesure des _tier list_ ; Mario a toujours été mauvais, mais il a eu un pic avant de stagner...
